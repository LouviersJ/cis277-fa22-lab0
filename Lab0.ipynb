## While Loop (less Pythonic)

```python
s = "RacecAr "

def is_palindrome1(string):
    clean_string = string.lower().replace(' ', '')
    first, last = 0, len(clean_string)-1
    while( first < last):
        if clean_string[first] == clean_string[last]:
            first +=1
            last -=1
        else:
            return False
    
    return True

is_palindrome1(s) 
```

## String Slicing (More Pythonic)

```python
def is_palindrome2(string):
    clean_string = string.lower().replace(' ', '')
    reversed_string = clean_string[::-1]
    return clean_string == reversed_string

s = " Rac eC ar "
is_palindrome2(s)
```

```python
import csv

movie = {}


with open('movie.csv', encoding="utf8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #print(row['movie_title'], row['imdb_score'])
        movie[row['movie_title']] = [float(row['imdb_score']), row['director_name'], row['duration'], row['genres']]

movie
```

What is the average IMDB Score of all the movies?

```python
sum = 0

for mov in movie.values():
    sum += mov[0]

average_rating = sum / len(movie)

print(average_rating)
```
